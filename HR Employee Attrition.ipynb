# HR Employee Attrition

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Employee Atrition.csv')
```

```python
data = data.drop(columns=['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'])
```

# Attrition ratas analysis

```python
sns.countplot(x='Attrition', data=data)
plt.title('Attrition Distribution')
plt.show()
```

# Attrition by Department

```python
sns.countplot(x='Department', hue='Attrition', data=data)
plt.title('Attrition by Department')
plt.show()
```

# Attrition by Age

```python
plt.figure(figsize=(8,6))
sns.histplot(data=data, x='Age', hue='Attrition', multiple='stack')
plt.title('Attrition by Age')
plt.show()
```

# Attrition by Gender

```python
sns.countplot(x='Gender', hue='Attrition', data=data)
plt.title('Attrition by Gender')
plt.show()
```

# Attrition by Job Role

```python
plt.figure(figsize=(10,6))
sns.countplot(y='JobRole', hue='Attrition', data=data)
plt.title('Attrition by Job Role')
plt.show()
```

# Monthly Income vs. Attrition

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=data)
plt.title('Monthly Income vs Attrition')
plt.show()
```

# Years at Company vs. Attrition

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Attrition', y='YearsAtCompany', data=data)
plt.title('Years at Company vs Attrition')
plt.show()
```

# Job Satisfaction vs. Attrition 

```python
plt.figure(figsize=(8,6))
sns.countplot(x='JobSatisfaction', hue='Attrition', data=data)
plt.title('Job Satisfaction vs Attrition')
plt.show()
```

# Attrition by Job Level

```python
plt.figure(figsize=(8,6))
sns.countplot(x='JobLevel', hue='Attrition', data=data)
plt.title('Attrition by Job Level')
plt.show()
```

## Suggestions:
Improve Work-Life Balance:
Employees with poor work-life balance are more likely to leave the company. Implementing flexible working hours, remote work options, and wellness programs can help improve work-life balance and reduce attrition.

Enhance Job Satisfaction:
Regular feedback sessions, recognition programs, and career development opportunities can boost job satisfaction. Satisfied employees are less likely to leave, so focusing on these areas is crucial.

Competitive Compensation:
As seen in the analysis, lower income levels are associated with higher attrition rates. Offering competitive salaries and regular salary hikes can help retain talent, especially in lower job levels.

Focus on Career Growth:
Employees at lower job levels are more prone to leaving. Providing clear career paths, promotions, and skill development opportunities can motivate them to stay and grow within the company.

Targeted Retention Strategies:
Different job levels and departments may require tailored retention strategies. HR should analyze specific needs and challenges of each group to implement effective measures.


## Conclusion:
The project has provided valuable insights into the factors influencing employee attrition. By understanding the relationships between job satisfaction, work-life balance, income, and attrition, the company can make informed decisions to enhance employee retention.

Implementing the suggested strategies will not only help reduce turnover but also improve overall employee engagement, satisfaction, and productivity. Continuous monitoring and analysis of these factors will allow the company to stay ahead of potential attrition issues and maintain a motivated and committed workforce.

In conclusion, the data-driven approach in this project has laid the foundation for developing effective HR strategies that align with the needs and expectations of employees, ensuring long-term success for both the employees and the organization.
