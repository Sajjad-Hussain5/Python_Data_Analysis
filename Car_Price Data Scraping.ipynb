```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

url = "https://www.pakwheels.com/used-cars/"
headers = {'User-Agent': 'Mozilla/5.0'}
webpage = requests.get(url, headers=headers)

soup = BeautifulSoup(webpage.content, "html.parser")

Cars_Name = []
Prices = []
Cities = []

for d in soup.find_all('div', attrs={'class': 'cards-content'}):
    Car_Name = d.find('a').text.strip()  # Using d instead of soup
    Price = d.find('div', class_='generic-green').text.strip()  # Using d instead of soup
    City = d.find('div', class_='generic-gray').text.strip()  # Using d instead of soup
    
    Cars_Name.append(Car_Name)
    Prices.append(Price)
    Cities.append(City)

print(Cars_Name)
print(Prices)
print(Cities)

df = pd.DataFrame({
    'Car Name': Cars_Name,
    'Price': Prices,
    'City': Cities
})

# Exporting the DataFrame to a CSV file
df.to_csv('used_cars.csv', index=False)

print("Data has been saved to 'used_cars.csv'.")
```
